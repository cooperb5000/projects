```python
df = pd.read_csv('decliners.csv')
df.Date = pd.to_datetime(df.Date)
df1 = df.groupby(['Landing Page', pd.Grouper(key='Date', freq='30D', origin='end')])['Url Clicks'].sum()

df1 = pd.DataFrame(df1)
df1 = df1.pivot_table(
        values='Url Clicks', 
        index=['Landing Page'], 
        columns='Date', 
        aggfunc=np.sum)

months = df1.columns
months = sorted(months)
df3 = df1[months]


df3.reset_index(inplace=True)
df3.rename(columns={'Landing Page': 'url'}, inplace=True)

df1 = pd.read_csv('allposts.csv', usecols= ['permalink', 'segment', 'original_pub_date'])
df1.rename(columns={'permalink': 'url'}, inplace=True)
gbr = 'https://www.gobankingrates.com'
df1['url'] = gbr + df1['url'].astype('string')

df3 = pd.merge(df3, df1, how='inner')
df3 = df3.drop_duplicates(subset='url')
df3.reset_index(inplace=True)

df2 = df3.drop(df3.columns[[-1, -2]], axis=1)
df2.drop(columns=['index'], inplace=True)
df2
```

```python
def foo(pct1, pct2, pct3, pct4, pct5, pct6):
    growth_over1 = (pct2 - pct1)/(pct1)
    growth_over2 = (pct3 - pct2)/(pct2)
    growth_over3 = (pct4 - pct3)/(pct3)
    growth_over4 = (pct5 - pct4)/(pct4)
    growth_over5 = (pct6 - pct5)/(pct5)



    return growth_over5, growth_over4, growth_over3, growth_over2, growth_over1



#create a function so that we can compare growth percentages quickly & keep code cleaner
```

```python


m = df2.max(axis=1)
m = (list(m))

past_top = []
past_value = []
may_value = []
max_value = []
difference = []
past_decline_pct = []
segmentd = []



decliner = []
decliner_value= []
decliner_growth_pct = []
j_value = []
mx_value = []
d_value = []
segmentl = []


null = []

mnthf = df3[months[-1]]
mnthn = df3[months[-2]]
mnth6 = df3[months[6]]
mnth5 = df3[months[5]]
mnth4 = df3[months[4]]
mnth3 = df3[months[3]]
mnth2 = df3[months[2]]
mnth1 = df3[months[1]]


##make sure you switch back to 6 months

for i in range(0, len(m)):    

    #include may because even month decline would be interesting
    
    if ((mnthf[i]-m[i])/m[i] < -0.2) and (m[i] > 10000):
        past_top.append(df3['url'][i])
        past_value.append(mnthf[i])
        may_value.append(mnthn[i])
        max_value.append(m[i])
        difference.append(m[i]-mnthf[i])
        segmentd.append(df3['segment'][i])
        past_decline_pct.append(foo(mnth1[i], mnth2[i], mnth3[i], mnth4[i], mnth5[i], mnth6[i]))
    
   
    
    #checks if january traffic was over 5000, then jan --> may decline was greater than 10% & jan ---> june was greater than 20% 
    if ((mnthf[i]-m[i])/m[i] < -0.2) and (m[i] > 100):
       decliner.append(df3['url'][i])
       j_value.append(mnthf[i])
       decliner_value.append(mnthn[i])
       mx_value.append(m[i])
       d_value.append(m[i] - mnthn[i])
       segmentl.append(df3['segment'][i])
       decliner_growth_pct.append(foo(mnth1[i], mnth2[i], mnth3[i], mnth4[i], mnth5[i], mnth6[i]))
    
    #checks consistent decliners over last 3 months of at least 5%, clicks in jan must be over 100 
    else: 
        null.append(df3['url'][i])



# for loop to categorize data into correct bins

m

```

```python
past_top_performers = pd.DataFrame()

past_top_performers.insert(0, 'past_top', past_top)
past_top_performers.insert(1, '%s_clicks' % months[-2], may_value)
past_top_performers.insert(1, '%s_clicks' % months[-1], past_value)
past_top_performers.insert(1, 'Max_clicks', max_value)
past_top_performers.insert(4, 'Difference_Between_Max&_Current', difference)
past_top_performers.insert(5, 'Segment', segmentd)




decliners = pd.DataFrame()

decliners.insert(0, 'decliners', decliner)
decliners.insert(1, '%s_clicks' % months[-2], decliner_value)
decliners.insert(1, '%s_Clicks' % months[-1], j_value)
decliners.insert(1, 'Max_Clicks', mx_value)
decliners.insert(4, 'Difference_Between_Max&_Current', d_value)
decliners.insert(5, 'Segment', segmentl)



# decliners.insert(2, 'growth', decliner_growth_pct)
# decliners[['june_growth', 'may_growth', 'april_growth', 'mar_growth', 'feb_growth']] = (decliners['growth'].to_list())
# decliners.drop(['growth'], axis=1, inplace=True)


```

```python
past_top_performers.sort_values('Difference_Between_Max&_Current', ascending=False, inplace=True)
decliners.sort_values('Difference_Between_Max&_Current', ascending=False, inplace=True)

display(past_top_performers)
display(decliners)

past_top_performers.to_csv('past_top_deciners.csv')
decliners.to_csv('all_decliners.csv')
```

```python
dfq = pd.read_csv('page1.csv')


dfq.rename(columns={'Query': 'query','Date (ISO Year Week)': 'date', 'Average Position': 'position', 'Clicks': 'clicks'}, inplace=True)


dfq = dfq.groupby(['query', 'date'])[['position', 'clicks']].agg('sum', 'mean')
#group articles by url & include date to make sure no duplicates, agrgrate sum & mean

dfq = pd.DataFrame(dfq)

dfq.sort_values('date', ascending=False, inplace=True)
dfq.reset_index(inplace=True)


values = ['position', 'clicks']
dfq = dfq.pivot(index='query', columns='date', values=(values))
#pivot so it is organized how we want, URL, Date, Clicjs
columns = list(dfq.columns.values)
print(columns)
dfq.reset_index(inplace=True)

columns.insert(0, 'query')

dfq.columns=(columns)
dfq.sort_values(columns[-1], ascending=False, inplace=True)
display(dfq)




```

```python
n1 = dfq[columns[1]]
n2 = dfq[columns[2]]
c1 = dfq[columns[3]]
c2 = dfq[columns[4]]
q1 = dfq[columns[0]]
keyword = []
old_pos = []
new_pos = []
old_clicks = []
new_clicks = []
d = []
na = []

for s in range(0, len(dfq['query'])):
    if (n1[s]<10.999) and  (n2[s]>10.999): 
        keyword.append(q1[s])
        old_pos.append(n1[s])
        new_pos.append(n2[s])
        old_clicks.append(c1[s])
        new_clicks.append(c2[s])
        d.append(c2[s] - c1[s])
        
    else:
        na.append(q1[s])
        
```

```python

lost = pd.DataFrame()

lost.insert(0, 'query', keyword)
lost.insert(1, columns[1], old_pos)
lost.insert(2, columns[2], new_pos)
lost.insert(3, columns[3], old_clicks)
lost.insert(4, columns[4], new_clicks)
lost.insert(5, "Clicks_Difference", d)


lost.sort_values("Clicks_Difference", ascending=True, inplace=True)
display(lost)
```
